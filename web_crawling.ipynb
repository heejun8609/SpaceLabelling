```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd

# 파이어폭스 실행
driver = webdriver.Firefox()
```

```python
# 파이어폭스 창 뜨고 실행 사이트로 이동
driver.get("http://www.instagram.com")
```

```python
# 역 목록 가져오기
station = pd.read_excel("2호선 역 목록.xlsx")
# 역 이름에 # 붙여서 리스트로 저장
station = list(station['이예규'])
stations = [x for x in station if str(x) != 'nan']
stations
```

```python
for sta in stations[3:]:
    print(sta)
    driver.get("https://www.instagram.com/explore/tags/"+ sta)
    time.sleep(2)
    postCnt = driver.find_element_by_tag_name("header")
    postCnt = int(postCnt.find_element_by_tag_name("span").text.replace(",","").split()[0])
    print(postCnt)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight-50);")
    a_tag = driver.find_elements_by_tag_name("a")
    try:
        idx = [x.get_attribute("text") for x in a_tag].index("Load more")
        a_tag[idx].click()
        time.sleep(2)
    except:
        continue
    img = []
    latestimglen = 0
    while postCnt > len(img) and len(img)<=15000:
        img = driver.find_elements_by_tag_name("img")
        if latestimglen == len(img):
            break
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight-50);")
        time.sleep(2)
        print(len(img))
        latestimglen = len(img)
    img = driver.find_elements_by_tag_name("img")
    img_tag = [x.get_attribute("alt").replace("\n", "") for x in img]
    print(len(img_tag), "개")
    pd.DataFrame(img_tag).to_csv(sta + ".csv", header=False, index=False)
```
